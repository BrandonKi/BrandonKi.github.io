```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(0, 2 * np.pi, 100)
z = np.linspace(0, 7, 50)
theta_grid, z_grid = np.meshgrid(theta, z)
r = np.sqrt(5)
x_grid = r * np.cos(theta_grid)
y_grid = r * np.sin(theta_grid)

ax.plot_surface(x_grid, y_grid, z_grid, color='green', alpha=0.6)

width = r + 1
x_plane = np.linspace(-width, width, 50)
y_plane = np.linspace(-width, width, 50)
x_plane, y_plane = np.meshgrid(x_plane, y_plane)
z_plane = 3 - x_plane - y_plane

ax.plot_surface(x_plane, y_plane, z_plane, color='orange', alpha=0.5)

x_intersection = np.sqrt(5) * np.cos(theta)
y_intersection = np.sqrt(5) * np.sin(theta)
z_intersection = 3 - x_intersection - y_intersection

ax.plot(x_intersection, y_intersection, z_intersection, color='red', linewidth=3)

# Set the labels and title
# ax.set_xlabel('X axis')
# ax.set_ylabel('Y axis')
# ax.set_zlabel('Z axis')
# ax.set_title('Intersection of 3D Cylinder x^2 + y^2 = 5 and Shifted Slanted Plane')

ax.set_axis_off()  # Hide the axes
ax.set_title('')    # Remove the title

# Display the plot
plt.show()
```
